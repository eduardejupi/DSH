```python
# ============================================================
# LABORATORIO 6 - ANALISI DA FILE ESISTENTI (NUOVO NOTEBOOK)
# ============================================================

# 0. IMPORT LIBRERIE
# ------------------------------------------------------------
from google.colab import drive
drive.mount('/content/drive')

import os
import ast
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# ============================================================
# 1. DEFINISCI PERCORSI FILE (ADATTA ALLA TUA STRUTTURA)
# ------------------------------------------------------------
# Modifica qui se i file sono in un'altra cartella:
base_path = "/content/drive/MyDrive/Lab-gruppo/Lab5"

train_path = os.path.join(base_path, "all_models_training_metrics.csv")
val_path   = os.path.join(base_path, "all_models_validation_metrics.csv")
test_path  = os.path.join(base_path, "all_models_test_metrics.csv")
fs_cls_path = os.path.join(base_path, "FS_CLASSIFIER_RESULTS_TRS_VS_TUNED-2.csv")  # se esiste
mask_comp_path = os.path.join(base_path, "mask_comparison_metrics.csv")

# ============================================================
# 2. CARICAMENTO DEI CSV (LAB 4–5)
# ------------------------------------------------------------
print("Carico i file metrics...")

train_df = pd.read_csv(train_path)
val_df   = pd.read_csv(val_path)
test_df  = pd.read_csv(test_path)

print("Shape train:", train_df.shape)
print("Shape val  :", val_df.shape)
print("Shape test :", test_df.shape)

if os.path.exists(fs_cls_path):
    fs_cls_df = pd.read_csv(fs_cls_path)
    print("Caricato FS_CLASSIFIER_RESULTS_TRS_VS_TUNED-2.csv:", fs_cls_df.shape)
else:
    fs_cls_df = None
    print("⚠️ File FS_CLASSIFIER_RESULTS_TRS_VS_TUNED-2.csv non trovato (non è essenziale).")

# ============================================================
# 3. RIASSUNTO PRESTAZIONI TRS / VS / TS
# ------------------------------------------------------------
# Rinominazione per evitare collisioni
train_tmp = train_df.rename(columns={
    "Balanced Accuracy": "BA_Train",
    "ROC AUC": "AUC_Train"
})
val_tmp = val_df.rename(columns={
    "Balanced Accuracy": "BA_Val",
    "ROC AUC": "AUC_Val"
})
test_tmp = test_df.rename(columns={
    "Balanced Accuracy": "BA_Test",
    "ROC AUC": "AUC_Test"
})

# Merge su 'Model'
summary_df = (
    train_tmp[["Model", "BA_Train", "AUC_Train"]]
    .merge(val_tmp[["Model", "BA_Val", "AUC_Val"]], on="Model", how="inner")
    .merge(test_tmp[["Model", "BA_Test", "AUC_Test"]], on="Model", how="inner")
)

# Calcolo GAP (overfitting approx)
summary_df["Gap_Train_Val"] = summary_df["BA_Train"] - summary_df["BA_Val"]

print("\n=== RIASSUNTO COMPLESSIVO (prime righe) ===")
display(summary_df.head())

# ============================================================
# 4. FILTRO MODELLI CON TROPPO OVERFITTING
# ------------------------------------------------------------
# Criterio: GAP <= 0.15 (Train non molto più alto di Val)
gap_threshold = 0.15
filtered_models = summary_df[summary_df["Gap_Train_Val"] <= gap_threshold].copy()
overfit_models  = summary_df[summary_df["Gap_Train_Val"] > gap_threshold].copy()

print("\n=== MODELLI FILTRATI PER OVERFITTING (Gap > 0.15) ===")
display(overfit_models.sort_values(by="Gap_Train_Val", ascending=False)[
    ["Model", "BA_Train", "BA_Val", "Gap_Train_Val"]
])

print("\n=== MODELLI 'BUONI' (Gap <= 0.15), ordinati per AUC_Val e BA_Test ===")
filtered_sorted = filtered_models.sort_values(
    by=["AUC_Val", "BA_Test"], ascending=[False, False]
)
display(filtered_sorted[["Model", "BA_Train", "BA_Val", "BA_Test", "AUC_Val", "AUC_Test", "Gap_Train_Val"]])

# Puoi scegliere da qui i tuoi "top modelli"
top_n = 8
top_models = filtered_sorted.head(top_n)["Model"].tolist()
print(f"\nTop {top_n} modelli (da usare per la discussione):")
for m in top_models:
    print(" -", m)

# ============================================================
# 5. CARICA RISULTATI DI CONFRONTO MASCHERE (Lab 6 - Step 1)
# ------------------------------------------------------------
print("\nCarico mask_comparison_metrics.csv...")
mask_comp_df = pd.read_csv(mask_comp_path)
print("Shape mask_comp_df:", mask_comp_df.shape)

print("\nEsempio righe mask_comparison_metrics:")
display(mask_comp_df.head())

# Controllo che ci siano le due maschere
print("Valori unici in 'Mask':", mask_comp_df["Mask"].unique())

# ============================================================
# 6. COSTRUZIONE TABELLA WIDE: METRICHE PER MODELLO E MASCHERA
# ------------------------------------------------------------
# Qui ci concentriamo soprattutto sulla Balanced Accuracy, ma teniamo anche AUC ecc.

# Pivot per Balanced Accuracy
ba_pivot = mask_comp_df.pivot_table(
    index="Model",
    columns="Mask",
    values="Balanced Accuracy"
)

# Rinomino colonne per chiarezza
ba_pivot = ba_pivot.rename(columns={
    "Mask 1": "BA_Mask1",
    "Mask 2": "BA_Mask2"
})

# Pivot per AUC (opzionale, se presente)
if "ROC AUC" in mask_comp_df.columns:
    auc_pivot = mask_comp_df.pivot_table(
        index="Model",
        columns="Mask",
        values="ROC AUC"
    ).rename(columns={
        "Mask 1": "AUC_Mask1",
        "Mask 2": "AUC_Mask2"
    })
else:
    auc_pivot = None

# Merge BA + AUC (se disponibile)
mask_summary = ba_pivot.copy()
if auc_pivot is not None:
    mask_summary = mask_summary.merge(auc_pivot, on="Model", how="left")

# Calcola Delta BA e |Delta BA|
mask_summary["Delta_BA"] = mask_summary["BA_Mask2"] - mask_summary["BA_Mask1"]
mask_summary["Abs_Delta_BA"] = mask_summary["Delta_BA"].abs()

# Ordina per stabilità (variazione più piccola)
mask_summary_sorted = mask_summary.sort_values(by="Abs_Delta_BA", ascending=True)

print("\n=== Variazione Balanced Accuracy per modello (Mask2 - Mask1), ordinato per |ΔBA| ===")
display(mask_summary_sorted.head(20))

# Salva la tabella (opzionale)
out_delta_path = os.path.join(base_path, "mask_delta_BA_per_model.csv")
mask_summary_sorted.to_csv(out_delta_path, index=True)
print(f"✅ Tabella per-model Delta BA salvata in: {out_delta_path}")

# ============================================================
# 7. SELEZIONA MODELLI ROBUSTI E NON ROBUSTI
# ------------------------------------------------------------
# Esempio: robusti se |ΔBA| <= 0.05, poco robusti se |ΔBA| >= 0.20

robust_threshold = 0.05
nonrobust_threshold = 0.20

robust_models = mask_summary_sorted[mask_summary_sorted["Abs_Delta_BA"] <= robust_threshold]
nonrobust_models = mask_summary_sorted[mask_summary_sorted["Abs_Delta_BA"] >= nonrobust_threshold]

print(f"\n=== MODELLI PIÙ ROBUSTI (|ΔBA| <= {robust_threshold}) ===")
display(robust_models)

print(f"\n=== MODELLI POCO ROBUSTI (|ΔBA| >= {nonrobust_threshold}) ===")
display(nonrobust_models)

# ============================================================
# 8. GRAFICO SCATTER: BA_Mask1 vs BA_Mask2 PER TUTTI I MODELLI
# ------------------------------------------------------------
plt.figure(figsize=(8, 8))
plt.scatter(mask_summary["BA_Mask1"], mask_summary["BA_Mask2"], alpha=0.7)
plt.plot([0, 1], [0, 1], "--", color="gray")  # diagonale
plt.xlabel("Balanced Accuracy - Mask 1")
plt.ylabel("Balanced Accuracy - Mask 2")
plt.title("Confronto Balanced Accuracy per modello\nMask 1 vs Mask 2")
plt.grid(True)
plt.show()

# ============================================================
# 9. BARPLOT PER I TOP MODELLI (MASK 1 vs MASK 2)
# ------------------------------------------------------------
# Prendiamo l'intersezione fra i top (per prestazioni) e quelli presenti in mask_summary
models_to_plot = [m for m in top_models if m in mask_summary.index]

if len(models_to_plot) == 0:
    print("⚠️ Nessuno dei top_models trovato in mask_summary. Uso i primi N di mask_summary_sorted.")
    models_to_plot = mask_summary_sorted.index.tolist()[:8]

plot_df = mask_summary.loc[models_to_plot].reset_index()

# Per fare un barplot comparativo, portiamo i dati in formato 'long'
ba_long = plot_df.melt(
    id_vars="Model",
    value_vars=["BA_Mask1", "BA_Mask2"],
    var_name="Mask",
    value_name="Balanced Accuracy"
)

# Rinominazione maschere per leggibilità
ba_long["Mask"] = ba_long["Mask"].replace({
    "BA_Mask1": "Mask 1",
    "BA_Mask2": "Mask 2"
})

plt.figure(figsize=(12, 6))
sns.barplot(
    data=ba_long,
    x="Model",
    y="Balanced Accuracy",
    hue="Mask"
)
plt.title("Balanced Accuracy per modello: confronto Mask 1 vs Mask 2")
plt.ylim(0, 1.05)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Maschera")
plt.tight_layout()
plt.show()

# ============================================================
# 10. (OPZIONALE) GRAFICO DEL DELTA_BA PER MODELLO
# ------------------------------------------------------------
plt.figure(figsize=(12, 6))
delta_plot = mask_summary_sorted.reset_index()
sns.barplot(
    data=delta_plot,
    x="Model",
    y="Delta_BA"
)
plt.axhline(0, color="black", linewidth=1)
plt.title("ΔBalanced Accuracy (Mask 2 - Mask 1) per modello")
plt.xticks(rotation=90)
plt.ylabel("Delta_BA")
plt.tight_layout()
plt.show()

print("\n✅ Analisi Lab 6 - Step 1 completata.")
print("Usa le tabelle 'mask_delta_BA_per_model.csv' e i grafici per scrivere la sezione di risultati.")
```

```python
# ============================================================
# Calcolo DICE per tutte le maschere e categorizzazione
# ============================================================

from google.colab import drive
drive.mount('/content/drive')

import os
import numpy as np
import pandas as pd
import nibabel as nib   # per leggere maschere .nii / .nii.gz

# ====== 1) IMPOSTA QUI LE TUE CARTELLE ======
mask1_dir = "/content/drive/MyDrive/Lab-gruppo/Dataset_1/Maschere"  # es. Radiologo 1
mask2_dir = "/content/drive/MyDrive/Lab-gruppo/Dataset_2/Maschere"  # es. Radiologo 2

output_csv = "/content/drive/MyDrive/Lab-gruppo/DICE_masks_with_categories.csv"

# ====== 2) FUNZIONE PER IL DICE ======
def dice_coefficient(mask1, mask2):
    """
    mask1, mask2: array booleani (True = voxel dentro la maschera)
    DICE = 2 * |A ∩ B| / (|A| + |B|)
    """
    vol1 = mask1.sum()
    vol2 = mask2.sum()

    # Caso borderline: entrambe vuote
    if vol1 + vol2 == 0:
        return np.nan  # oppure 1.0 se vuoi considerarle "perfettamente uguali"

    intersection = np.logical_and(mask1, mask2).sum()
    return 2.0 * intersection / (vol1 + vol2)

# ====== 3) FUNZIONE PER CATEGORIA DICE ======
def dice_category(dice_value):
    """
    Assegna la categoria in base al valore di DICE.

    High DICE   : DICE > 0.65
    Medium DICE : 0.45 <= DICE <= 0.65
    Low DICE    : DICE < 0.45
    """
    if np.isnan(dice_value):
        return "NaN (maschere vuote)"
    if dice_value > 0.65:
        return "High DICE"
    elif dice_value >= 0.45:
        return "Medium DICE"
    else:
        return "Low DICE"

# ====== 4) TROVO I FILE IN COMUNE TRA LE DUE CARTELLE ======
files1 = {f: os.path.join(mask1_dir, f) for f in os.listdir(mask1_dir) if f.endswith((".nii", ".nii.gz"))}
files2 = {f: os.path.join(mask2_dir, f) for f in os.listdir(mask2_dir) if f.endswith((".nii", ".nii.gz"))}

common_files = sorted(set(files1.keys()) & set(files2.keys()))
print(f"Trovati {len(common_files)} file in comune tra le due cartelle.")

results = []

# ====== 5) LOOP SU TUTTI I PAZIENTI / FILE IN COMUNE ======
for fname in common_files:
    path1 = files1[fname]
    path2 = files2[fname]

    try:
        img1 = nib.load(path1).get_fdata()
        img2 = nib.load(path2).get_fdata()
    except Exception as e:
        print(f"⚠ Errore nel leggere {fname}: {e}")
        results.append({
            "filename": fname,
            "dice": np.nan,
            "dice_category": "Errore lettura"
        })
        continue

    # Controllo shape
    if img1.shape != img2.shape:
        print(f"⚠ Shape diversa per {fname}: {img1.shape} vs {img2.shape}")
        results.append({
            "filename": fname,
            "dice": np.nan,
            "dice_category": "Shape mismatch"
        })
        continue

    # Binarizzo le maschere (True = voxel > 0)
    mask1 = img1 > 0
    mask2 = img2 > 0

    d = dice_coefficient(mask1, mask2)
    cat = dice_category(d)

    results.append({
        "filename": fname,
        "dice": d,
        "dice_category": cat
    })

# ====== 6) CREO DATAFRAME E SALVO ======
df_dice = pd.DataFrame(results)
df_dice = df_dice.sort_values(by="dice", ascending=False)

display(df_dice.head())

df_dice.to_csv(output_csv, index=False)
print(f"\n✅ File DICE salvato in: {output_csv}")
```
