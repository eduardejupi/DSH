```python
import os, glob, re, pandas as pd
from radiomics import featureextractor

IN_DIR = "/content/drive/MyDrive/DSH/Lab-gruppo/dataset1_CS"
MASK_DIR = "/content/drive/MyDrive/DSH/Lab-gruppo/Dataset_2/Maschere"
PARAMS = "/content/drive/MyDrive/DSH/Lab-gruppo/Lab2/params.yaml"

# Inizializza l’estrattore (legge params.yaml)
extractor = featureextractor.RadiomicsFeatureExtractor(PARAMS)

rows = []
# Ordina per id numerico
mask_paths = sorted(
    glob.glob(os.path.join(MASK_DIR, "mask_*.nii.gz")),
    key=lambda p: int(re.findall(r"\d+", os.path.basename(p))[0])
)

for mask_path in mask_paths:
    k = int(re.findall(r"\d+", os.path.basename(mask_path))[0])
    img_path = os.path.join(IN_DIR, f"image_{k}.nii.gz")
    if not os.path.exists(img_path):
        print(f"Salto ID {k}: immagine non trovata")
        continue

    result = extractor.execute(img_path, mask_path)
    # Tieni solo le feature (escludi diagnosi/diagnostics)
    feats = {k2: v2 for k2, v2 in result.items() if not k2.startswith("diagnostics")}
    feats["ID"] = k
    rows.append(feats)

df_feats = pd.DataFrame(rows).sort_values("ID").reset_index(drop=True)

# Reorder columns to put 'ID' first
cols = df_feats.columns.tolist()
cols.insert(0, cols.pop(cols.index('ID')))
df_feats = df_feats[cols]

df_feats.to_csv("/content/drive/MyDrive/DSH/Lab-gruppo/Lab2/features_dataset2.csv", index=False)
print("✅ Salvato: /content/drive/MyDrive/DSH/Lab-gruppo/Lab2/features_dataset2.csv")
display(df_feats.head())
```

```python
from google.colab import auth
import gspread
from google.auth import default

# Authenticate to Google Drive and Google Sheets
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)

# Open the spreadsheet and select the worksheet
try:
    spreadsheet = gc.open("Diagnosi")
    worksheet = spreadsheet.get_worksheet(0)
    # Get all data from the worksheet
    diagnosis_data = worksheet.get_all_records()
    df_diagnosis = pd.DataFrame(diagnosis_data)
    print("✅ Caricato il file Diagnosi.gsheet")
    display(df_diagnosis.head())
except Exception as e:
    print(f"Error loading Google Sheet: {e}")
    df_diagnosis = None
```

```python
if df_diagnosis is not None:
    # Rename 'CASE NUMBER' to 'ID' for merging
    df_diagnosis = df_diagnosis.rename(columns={"CASE NUMBER": "ID"})

    # Select only the relevant columns and rename 'DEFINITIVE DIAGNOSIS' to 'class'
    df_diagnosis = df_diagnosis[['ID', 'DEFINITIVE DIAGNOSIS ']].rename(columns={'DEFINITIVE DIAGNOSIS': 'class'})

    # Ensure 'ID' column is of the same type in both dataframes for merging
    #df_diagnosis['ID'] = df_diagnosis['ID'].astype(str)
    #df_feats['ID'] = df_feats['ID'].astype(str)

    # Merge the dataframes
    df_merged = pd.merge(df_feats, df_diagnosis, on='ID', how='left')

    # Save the merged dataframe
    output_csv_path_merged = "/content/drive/MyDrive/DSH/Lab-gruppo/Lab2/features_dataset2.csv"
    df_merged.to_csv(output_csv_path_merged, index=False)

    print(f"✅ Salvato il file unito con la colonna 'class': {output_csv_path_merged}")
    display(df_merged.head())
else:
    print("Skipping merge as diagnosis data could not be loaded.")
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd # Import pandas if not already imported

if df_merged is not None:
    # Separate features and the class column
    features = df_merged.drop(columns=['ID', 'DEFINITIVE DIAGNOSIS ']) # Dropping the original class column as well
    class_col = df_merged[['ID', 'DEFINITIVE DIAGNOSIS ']].rename(columns={'DEFINITIVE DIAGNOSIS ': 'class'})

    # Min-Max Scaling
    scaler_minmax = MinMaxScaler()
    features_normalized = scaler_minmax.fit_transform(features)
    df_normalized = pd.DataFrame(features_normalized, columns=features.columns)
    df_normalized = pd.concat([class_col, df_normalized], axis=1) # Add ID and class back

    # Reorder columns for df_normalized
    cols_normalized = df_normalized.columns.tolist()
    cols_normalized.insert(0, cols_normalized.pop(cols_normalized.index('ID')))
    cols_normalized.append(cols_normalized.pop(cols_normalized.index('class')))
    df_normalized = df_normalized[cols_normalized]


    # Z-score Standardization
    scaler_zscore = StandardScaler()
    features_standardized = scaler_zscore.fit_transform(features)
    df_standardized = pd.DataFrame(features_standardized, columns=features.columns)
    df_standardized = pd.concat([class_col, df_standardized], axis=1) # Add ID and class back

    # Reorder columns for df_standardized
    cols_standardized = df_standardized.columns.tolist()
    cols_standardized.insert(0, cols_standardized.pop(cols_standardized.index('ID')))
    cols_standardized.append(cols_standardized.pop(cols_standardized.index('class')))
    df_standardized = df_standardized[cols_standardized]


    # Save to CSV files
    output_csv_path_normalized = "/content/drive/MyDrive/DSH/Lab-gruppo/Lab2/features_dataset2_normalized.csv"
    df_normalized.to_csv(output_csv_path_normalized, index=False)
    print(f"✅ Salvato il file normalizzato: {output_csv_path_normalized}")

    output_csv_path_standardized = "/content/drive/MyDrive/DSH/Lab-gruppo/Lab2/features_dataset2_standardized.csv"
    df_standardized.to_csv(output_csv_path_standardized, index=False)
    print(f"✅ Salvato il file standardizzato: {output_csv_path_standardized}")

else:
    print("Skipping scaling and saving as merged data could not be loaded.")
```
